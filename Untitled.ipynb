# Define Haversine Function for mathematical calculation of distance between 2 coordinate pairs

```python
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):

    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return round(c * r,1)
```

```python
# read 911 calls csv
initial_df = pd.read_csv("911.csv")
initial_df.head()
```

## Clean the 911 call Data

```python
# drop useless columns
initial_df=initial_df.drop(columns=['desc','twp', 'addr','e'])
initial_df.head()
```

```python
# Split 'title' into Department and level 2
initial_df[['Dept','Cat 2']] = initial_df['title'].str.split(': ',expand=True)
# drop the title column
initial_df=initial_df.drop(columns=['title'])
initial_df.head()
```

```python
# Split timestamp into date and time
initial_df[['Date','Time']] = initial_df['timeStamp'].str.split(' ',expand=True)

# grab the first two characters of the "Time" to have "hour of the day"
initial_df["Hour of Day"]=initial_df["Time"].str[:2]

# use the timestamp to get day of the week, then grab the first 3 characters for a short version
initial_df["Mth"]=pd.to_datetime(initial_df['timeStamp']).dt.month_name().str[:3]
initial_df["Date2"]=pd.to_datetime(initial_df['timeStamp']).dt.strftime('%d/%m/%Y')


# use the date and take the 3 and 4th characters to get Yr
initial_df["Yr"]=initial_df['Date'].str[2:4]

# use the timestamp to get day of the week, then grab the first 3 characters for a short version
initial_df["WeekDay"]=pd.to_datetime(initial_df['timeStamp']).dt.day_name().str[:3]

# use the timestamp to get day of the week by number
initial_df["WeekDay#"]=pd.to_datetime(initial_df['timeStamp']).dt.dayofweek

# Reorder Columns
initial_df=initial_df[["Dept","Cat 2","Yr","Mth","WeekDay","WeekDay#","Date","Date2","Hour of Day","Time","timeStamp","lat","lng","zip"]]
initial_df.head()
```

### Determine the Date Range for the 911 Call data

```python
# get the min and max date for the sample
min_date=initial_df["Date"].min()
min_date_df = initial_df[initial_df["Date"]==min_date].iloc[0]
min_date_value=f"{min_date_df['Mth']}-{min_date_df['Yr']}"

max_date=initial_df["Date"].max()
max_date_df = initial_df[initial_df["Date"]==max_date].iloc[0]
max_date_value=f"{max_date_df['Mth']}-{max_date_df['Yr']}"
```

```python
# Add Distance
initial_df['km from Philly'] = initial_df.apply(lambda row: haversine(lon1, 
                                            lat1, 
                                            row['lng'], 
                                            row['lat']), axis=1)
initial_df
```

```python

empty_df = pd.DataFrame(columns=['Low Q','Up Q',"IQR","L Bnd","U Bnd","#O-Liers"])

values = initial_df['km from Philly']
   
quartiles = values.quantile([.25,.5,.75])
lowerq = round(quartiles[0.25],2)
upperq = round(quartiles[0.75],2)

iqr = round(upperq - lowerq,2)

lower_bound = round(lowerq - (1.5*iqr),2)
upper_bound = round(upperq + (1.5*iqr),2)
    
outliers_count = (values.loc[(initial_df['km from Philly'] > upper_bound) | (initial_df['km from Philly'] < lower_bound)]).count()
    
data = {
            "Low Q":lowerq,
            "Up Q":upperq,
            "IQR":iqr,
            "L Bnd":lower_bound,
            "U Bnd":upper_bound,
            "#O-Liers":outliers_count}
    
empty_df.loc[1]=data
    
empty_df
```
